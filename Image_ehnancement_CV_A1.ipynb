## APPLYING ENHANCEMENT FILTERS

## **Averaging Filter**
It is the type of low-pass filter that is used to smooth images and remove noise.

```python
kernel_size =(4,4)
#applying averaging filter on grayscale images
av1 = cv2.blur(gray1,kernel_size)
av2 = cv2.blur(gray2,kernel_size)
av3 = cv2.blur(gray3,kernel_size)
av4 = cv2.blur(gray4,kernel_size)
av5 = cv2.blur(gray5,kernel_size)
av6 = cv2.blur(gray6,kernel_size)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(av1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(av2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(av3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(av4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(av5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(av6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')

plt.show()
```

## **Median Filter**
It is a non-linear filter used to remove salt-and-pepper noise from image.

```python
#on images get after averaging filter, median filter is applied
m1 = cv2.medianBlur(av1, 5)
m2 = cv2.medianBlur(av2, 5)
m3 = cv2.medianBlur(av3, 5)
m4 = cv2.medianBlur(av4, 5)
m5 = cv2.medianBlur(av5, 5)
m6 = cv2.medianBlur(av6, 5)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(m1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(m2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(m3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(m4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(m5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(m6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')

plt.show()
```

## **GussianBlur Filter**
After median filter, Sharpening filter will be applied but first bluring the images using gaussain-blur filter. It works by convolving the images with a Gaussian kernel.

```python
kernel_size = (5,5)
sigma = 1.5 #randomly initialized
#on images get after median filter, Gaussian Blur Filter is applied
b1 = cv2.GaussianBlur(m1,kernel_size,sigma)
b2 = cv2.GaussianBlur(m2,kernel_size,sigma)
b3 = cv2.GaussianBlur(m3,kernel_size,sigma)
b4 = cv2.GaussianBlur(m4,kernel_size,sigma)
b5 = cv2.GaussianBlur(m5,kernel_size,sigma)
b6 = cv2.GaussianBlur(m6,kernel_size,sigma)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(b1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(b2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(b3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(b4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(b5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(b6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')
plt.show()
```

## **Sharpening Filter**
After gussian blur filter, cv2.addWeighted function used to subtract the blurred image from the median-filtered image and combine the result with the median-filtered image to produce the final sharpened image. It used to enhance the edges and details in the images

```python
amount = 2.0 #randomly chosen
#after gaussian blur filter appling sharpening filter on it
bf1 = cv2.addWeighted(m1, 1 + amount, b1, -amount, 0)
bf2 = cv2.addWeighted(m2, 1 + amount, b2, -amount, 0)
bf3 = cv2.addWeighted(m3, 1 + amount, b3, -amount, 0)
bf4 = cv2.addWeighted(m4, 1 + amount, b4, -amount, 0)
bf5 = cv2.addWeighted(m5, 1 + amount, b5, -amount, 0)
bf6 = cv2.addWeighted(m6, 1 + amount, b6, -amount, 0)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(bf1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(bf2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(bf3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(bf4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(bf5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(bf6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')

plt.show()
```

## **Histogram Equilization**
It is a widely used image enhancement technique that aims to improve the contrast and brightness of an image by redistributing the intensities of the image pixels.

```python
#applying histogram equilization filter on images get after sharpening filter
hist1 = cv2.equalizeHist(bf1)
hist2 = cv2.equalizeHist(bf2)
hist3 = cv2.equalizeHist(bf3)
hist4 = cv2.equalizeHist(bf4)
hist5 = cv2.equalizeHist(bf5)
hist6 = cv2.equalizeHist(bf6)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(hist1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(hist2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(hist3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(hist4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(hist5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(hist6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')
plt.show()
```

## **Gamma Correction Filter**
Gamma correction is a non-linear technique used to adjust the brightness and contrast of an image. It can be used to correct images that appear too dark or too bright.

```python
#after histogram euilization, gamma correction filter applied
gamma = 0.5 # Set the gamma value (experiment with different values)
gc1 = np.power(hist1, gamma)
gc2 = np.power(hist2, gamma)
gc3 = np.power(hist3, gamma)
gc4 = np.power(hist4, gamma)
gc5 = np.power(hist5, gamma)
gc6 = np.power(hist6, gamma)

# Normalize the image to 0-255 range
igc1 = cv2.normalize(gc1, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
igc2 = cv2.normalize(gc2, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
igc3 = cv2.normalize(gc3, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
igc4 = cv2.normalize(gc4, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
igc5 = cv2.normalize(gc5, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
igc6 = cv2.normalize(gc6, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(igc1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(igc2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(igc3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(igc4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(igc5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(igc6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')

plt.show()
```

## **Adaptive Filter**
An adaptive filter can be used to enhance the image while preserving its edges and details. It can adapt to the local properties of the image and provide better results than traditional filters.

```python
#In the end adaptive filter is applied

block_size = 15 # Set the block size (experiment with different values)
c = 3 # Set the constant value (experiment with different values)
iaf1 = cv2.adaptiveThreshold(igc1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
iaf2 = cv2.adaptiveThreshold(igc2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
iaf3 = cv2.adaptiveThreshold(igc3, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
iaf4 = cv2.adaptiveThreshold(igc4, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
iaf5 = cv2.adaptiveThreshold(igc5, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
iaf6 = cv2.adaptiveThreshold(igc6, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)

fig, axes = plt.subplots(2, 3)
axis = axes.ravel()
axis[0].imshow(iaf1, cmap='gray')
axis[0].set_title("Flower")
axis[1].imshow(iaf2, cmap='gray')
axis[1].set_title("Girl")
axis[2].imshow(iaf3, cmap='gray')
axis[2].set_title("House")
axis[3].imshow(iaf4, cmap='gray')
axis[3].set_title("Nature")
axis[4].imshow(iaf5, cmap='gray')
axis[4].set_title("Heart")
axis[5].imshow(iaf6, cmap='gray')
axis[5].set_title("Tiger")

axis[0].axis('off')
axis[1].axis('off')
axis[2].axis('off')
axis[3].axis('off')
axis[4].axis('off')
axis[5].axis('off')

plt.show()
```
